## Line plot

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
var = [ 1,2,3,4,5,6,6,7]
var_1 = [4,3,6,7,5,2,1,9]
data_1 = pd.DataFrame({'var':var, 'var_1':var_1})
sns.lineplot(x ="var", y = "var_1", data = data_1)
plt.show()
```
