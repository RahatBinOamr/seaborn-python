## Barplot

```python
import pandas as  pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
sns.set(style = 'darkgrid')
sns.barplot(penguins, x="island", y="body_mass_g", hue="sex", err_kws={'color': 'r'})
plt.show()
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
flights_wide = flights.pivot(index="year", columns="month", values="passengers")
sns.barplot(flights_wide,saturation=1)
plt.show()
```

```python
ax=sns.barplot(flights, x="year", y="passengers", estimator="sum", errorbar=None, native_scale=True,)
ax.plot(1955, 3600, "*", markersize=10, color="r")
plt.show()
```
